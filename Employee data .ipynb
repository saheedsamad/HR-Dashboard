```python
import pandas as pd

data2 = pd.read_csv('employee_data.csv')
data2
```

```python
import pandas as pd

# my first step is to drop the unnamec coloumn
columns_to_drop = ['Unnamed: 0']  
data2 = data2.drop(columns=columns_to_drop, errors='ignore')

# my second step i rename the dataset 
data2 = data2.rename(columns={
    'EmpID': 'EmployeeID',
    'FirstName': 'First Name',
    'LastName': 'Last Name',
    'StartDate': 'Start Date',
    'ExitDate': 'Exit Date',
    'ADEmail': 'Email',
    'BusinessUnit': 'Business Unit',
    'JobFunctionDescription': 'Job Function',
    'Performance_Score': 'Performance Score'
})

# my third step i fill missing data in exit column with still employed
data2['Exit Date'] = data2['Exit Date'].fillna('Still Employed')

# my fourth step i replace other missing values with 'Unknown' or 0
data2 = data2.fillna({'Supervisor': 'Unknown', 'Email': 'Unknown', 'Performance Score': 0})

# my fifth step Convert date columns to datetime
data2['Start Date'] = pd.to_datetime(data2['Start Date'], errors='coerce')
data2['Exit Date'] = pd.to_datetime(data2['Exit Date'], errors='coerce', format='%Y-%m-%d', exact=False)

# my step sixth step i Create additional calculated columns (tenure column)
data2['Tenure (Years)'] = data2.apply(
    lambda row: (pd.Timestamp.today() - row['Start Date']).days // 365 if pd.notna(row['Start Date']) else None,
    axis=1
)

# my seventh step i Ensure consistent formatting (tittle case for job functionand ensure numeric in performance score)
data2['Job Function'] = data2['Job Function'].str.title()  
data2['Performance Score'] = pd.to_numeric(data2['Performance Score'], errors='coerce')  

# my eight step (display the data ,display the first 5 row)
print(data2.info())
print(data2.head())

# my last step i save it in a different file to differentiate it from the original data 
data2.to_csv('Cleaned_Employee_Data.csv', index=False)

```
